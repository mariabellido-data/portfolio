```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv("../data/leadership_toxicity_kpis.csv")

# Estilo visual
sns.set_theme(style="whitegrid", context="talk")

# Grfico 1: distribucin de toxicidad
plt.figure(figsize=(8,5))
sns.histplot(df["supervisor_toxicity"], kde=True, color="red")
plt.title("Distribution of Supervisor Toxicity Levels")
plt.savefig("../reports/toxicity_distribution.png")
plt.show()
```

```python
# Correlacin entre variables clave
plt.figure(figsize=(6,4))
corr = df[["supervisor_toxicity","absenteeism_days","performance_score","motivation"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix: Toxicity, Absenteeism, Performance, Motivation")
plt.savefig("../reports/correlations_heatmap.png")
plt.show()

```

### Ethical Interpretation

This dashboard suggests that toxic supervision correlates strongly with:
- Higher absenteeism rates  
- Lower performance  
- Reduced motivation  

Ethical auditing of workplace performance means **translating data into accountability**.  
Numbers alone do not condemn  they **reveal systemic patterns** that leadership must address.
